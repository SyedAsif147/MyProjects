```python
def search():
    #connecting again
    params=config()
    conn=psycopg2.connect(**params)
    c=conn.cursor()
    #sql display
    def select():
        params=config()
        conn=psycopg2.connect(**params)
        c=conn.cursor()
        
        #selecting a record
        c.execute("select * from patients where patient_id = %s",(p_id.get(),))
        records=c.fetchone()
        print_rec=''
        m=3
        for i in records:
            disp_label=Label(root4,text = i)
            disp_label.grid(row=m,column=0,columnspan=2)
            m=m+1
        #clearing text boxes
        p_id.delete(0, tk.END)
        conn.commit()
        conn.close()
    #displaying all records
    root4=tk.Tk()
    root4.title("Record Search")
    p_id=Entry(root4,width=30)
    p_id.grid(row=0, column=1, padx=20)
    p_id_label=Label(root4,text="Patient ID")
    p_id_label.grid(row=0, column=0)
    search_btn=Button(root4,text="Show Record",command=select)
    search_btn.grid(row=2,column=0,columnspan=2,padx=20,pady=10,ipadx=100)
    root4.mainloop()
    #closing connection
    conn.commit()
    conn.close() 
```

```python
def display():
    #connecting again
    params=config()
    conn=psycopg2.connect(**params)
    c=conn.cursor()
    #sql display
    def disp():
        params=config()
        conn=psycopg2.connect(**params)
        c=conn.cursor()
        c.execute("select * from patients")
        records=c.fetchall()
        print_record=''
        for i in records:
            print_record+=str(i)+'\n'
        disp_label=Label(root5,text = print_record)
        disp_label.pack()
        conn.commit()
        conn.close()
    #displaying all records
    root5=tk.Tk()
    root5.title("Record Display")
    disp_btn=Button(root5,text="Show Records",command=disp)
    disp_btn.pack(side=TOP,padx=20,pady=10,ipadx=100)
    root5.mainloop()
    #closing connection
    conn.commit()
    conn.close()    
```

```python
def delete():
    counta=0
    countb=0
    #connecting again
    params=config()
    conn=psycopg2.connect(**params)
    c=conn.cursor()
    #sql display
    def delt():
        params=config()
        conn=psycopg2.connect(**params)
        c=conn.cursor()
        c.execute("select count(*) from patients")
        countb=c.rowcount
        #selecting a record
        c.execute("select * from patients where patient_id = %s",(p_id.get(),))
        idt=c.fetchone()
        c.execute("delete from patients where patient_id = %s",(p_id.get(),))
        counta=countb-1
        k=''
        for i in idt:
            k+=str(i)
            break
        txt="No of rows before deletion: "+str(countb)+"\nNo of rows after deletion: "+str(counta)
        txt2="Record with patient id "+str(k)+" has been deleted"
        del_label=Label(root6,text = txt)
        del_label.grid(row=5,column=0,columnspan=2)
        des_label=Label(root6,text = txt2)
        des_label.grid(row=4,column=0,columnspan=2)
        #clearing text boxes
        p_id.delete(0, tk.END)
        conn.commit()
        conn.close()
    #displaying all records
    root6=tk.Tk()
    root6.title("Record Deletion")
    p_id=Entry(root6,width=30)
    p_id.grid(row=0, column=1, padx=20)
    p_id_label=Label(root6,text="Patient ID")
    p_id_label.grid(row=0, column=0)
    search_btn=Button(root6,text="Delete Record",command=delt)
    search_btn.grid(row=2,column=0,columnspan=2,padx=20,pady=10,ipadx=100)
    root6.mainloop()
    #closing connection
    conn.commit()
    conn.close() 
```
